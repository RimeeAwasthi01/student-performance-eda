## Data Visualization

Visualizing relationships and patterns for better understanding of student performance.

### GPA Distribution

This histogram shows how GPA values are distributed among students.

```python
plt.figure(figsize=(6,4))
plt.hist(df["GPA"], bins=10)
plt.title("GPA Distribution")
plt.xlabel("GPA")
plt.ylabel("Number of Students")
plt.grid()
plt.show()
```

- Most students fall within a narrow GPA range.
- Very few students achieve high GPA, indicating overall moderate performance.

### Study Time vs GPA

This scatter plot shows the relationship between study hours and academic performance.

```python
plt.scatter(df["StudyTimeWeekly"], df["GPA"])
plt.title("Study Time vs GPA")
plt.xlabel("Study Time (Weekly)")
plt.ylabel("GPA")
plt.show()
```

- A positive trend is observed between study time and GPA.
- Students who study more tend to perform better.

### Absences vs GPA

This plot shows how attendance affects academic performance.

```python
plt.scatter(df["Absences"], df["GPA"])
plt.title("Absences vs GPA")
plt.xlabel("Absences")
plt.ylabel("GPA")
plt.show()
```

- A negative relationship is observed.
- Higher absences generally lead to lower GPA.

### Parental Support vs GPA

This bar chart shows how parental involvement affects student performance.

```python
df.groupby("ParentalSupport")["GPA"].mean().plot(kind="bar")
plt.title("Parental Support vs GPA")
plt.xlabel("Support Level")
plt.ylabel("Average GPA")
plt.show()
```

- GPA increases consistently with parental support.
- Strong parental involvement significantly improves performance.

### Activities vs GPA

This chart shows the impact of extracurricular activities.

```python
df.groupby("TotalActivities")["GPA"].mean().plot(kind="bar")
plt.title("Activities vs GPA")
plt.xlabel("Number of Activities")
plt.ylabel("Average GPA")
plt.show()
```

- Moderate participation improves GPA.
- Excessive involvement may slightly reduce performance.

## Key Findings

- Study time shows a positive impact on GPA.
- Absenteeism negatively affects performance.
- Parental support strongly improves academic outcomes.
- Moderate extracurricular involvement is beneficial.

## Conclusion

This analysis demonstrates that student performance is influenced by multiple interconnected factors including study habits, attendance, and support systems.

Students who maintain consistent study routines, have lower absenteeism, and receive strong parental support tend to achieve higher GPA.

The results also indicate that balanced extracurricular participation contributes positively, while excessive involvement may slightly hinder performance.

Overall, improving academic discipline and support structures can significantly enhance student outcomes.
